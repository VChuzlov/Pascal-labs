```python
x = np.linspace(0, 1, 1000)
actual = np.exp(np.tan(x) + 1)
```

```python
plt.style.use('ggplot')

fig, axs = plt.subplots(2, 1, figsize=(10, 7), sharex=True)

axs[0].plot(x, actual, label=r'Аналитическое решение $y=e^{tg(x)+1}$')
axs[0].plot(np.arange(0, 1.2, 0.2), eiler_02, marker='o', label='Метод Эйлера, h = 0.2')
axs[0].plot(np.arange(0, 1.1, 0.1), eiler_01, marker='o', label='Метод Эйлера, h = 0.1')
axs[0].plot(np.arange(0, 1.05, 0.05), eiler_005, marker='o', label='Метод Эйлера, h = 0.05')
axs[0].set_ylabel('Y')
axs[0].legend(fontsize=16)

axs[1].plot(x, actual, label=r'Аналитическое решение $y=e^{tg(x)+1}$')
axs[1].plot(np.arange(0, 1.2, 0.2), rk_02, marker='o', label='Метод Рунге-Кутты, h = 0.2')
axs[1].plot(np.arange(0, 1.1, 0.1), rk_01, marker='o', label='Метод Рунге-Кутты, h = 0.1')
axs[1].plot(np.arange(0, 1.05, 0.05), rk_005, marker='o', label='Метод Рунге-Кутты, h = 0.05')
axs[1].set_ylabel('Y')
axs[1].set_xlabel('X')
axs[1].legend(fontsize=16)

plt.tight_layout()
plt.savefig('eiler_vs_rk.svg')
```
