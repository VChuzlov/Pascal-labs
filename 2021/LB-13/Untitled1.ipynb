```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(0, 1.1, 0.1)
```

```python
ca_e = np.array([1,0.915,0.8381,0.7685,0.7055,0.6484,0.5968,0.5501,0.5079,0.4696,0.435, ])
```

```python
cb_e = np.array([0,0.085,0.1619,0.2315,0.2945,0.3516,0.4032,0.4499,0.4921,0.5304,0.565, ])
```

```python
ca_rk = np.array([1, 0.9189,0.8452,0.7781,0.7171,0.6617,0.6113,0.5654,0.5237,0.4858,0.4513, ])
```

```python
cb_rk = np.array([0,0.0811,0.1548,0.2219,0.2829,0.3383,0.3887,0.4346,0.4763,0.5142,0.5487, ])
```

```python
plt.style.use('ggplot')

fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

axs[0].plot(time, ca_e, marker='o', label=r'$C_A$')
axs[0].plot(time, cb_e, marker='o', label=r'$C_B$')
axs[0].set_title('Метод Эйлера')
axs[0].set_ylabel('Концентрация компонента, моль / л')
axs[0].set_xlabel('Время, с')
axs[0].legend()

axs[1].plot(time, ca_rk, marker='o', label=r'$C_A$')
axs[1].plot(time, cb_rk, marker='o', label=r'$C_B$')
axs[1].set_title('Метод Рунге-Кутты')
axs[1].set_xlabel('Время, с')
axs[1].legend()

plt.tight_layout()
plt.savefig('fig.svg')
```
